# SubPlots
A different way of plotting graph

```python
# batter dataset
batters.head()
```

```python
plt.figure(figsize = (15,6))
plt.scatter(batters['avg'],batters['strike_rate'],alpha =0.7)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
# title in this 
plt.title('Something')
```

```python
fig , ax = plt.subplots(figsize = (15,6))
ax.scatter( batters['avg'] , batters['strike_rate'] ,color='pink' , marker='+')
ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike')
```

```python
fig, ax = plt.subplots()
ax.plot(batters['avg'] , batters['strike_rate'])
ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike')
```

```python
# side by side plot
plt.figure(figsize = (15,20))
plt.subplots(nrows=2, ncols=2)
```

```python
fig, ax= plt.subplots(nrows=2, ncols=1 ,sharex=True, figsize= (10,6)) # sharex is when same on x axis it will show value in 
ax[0].scatter(batters['avg'],batters['strike_rate']) # below plot because they are shared
ax[1].scatter(batters['avg'],batters['runs'])
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strike rate')
ax[1].set_title('Average vs runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Run')
plt.plot()
```

```python
fig, ax= plt.subplots(nrows=2, ncols=2 , figsize= (10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate']) # below plot because they are shared
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg']) # below plot because they are shared
ax[1,1].hist(batters['runs'])
plt.plot()
```

```python
# 3rd way

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'])

ax2 = fig.add_subplot(222)
ax2.hist(batters['avg'])

ax3 = fig.add_subplot(223)
ax3.hist(batters['runs'],color ='blue')
```

# 3D surface plot
- scatter + surface together

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
ax = plt.figure(figsize=(12,8)).add_subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap ='viridis')
plt.colorbar(p) 
# yellow represent high z value
# purple represent low z value
```

```python
z = np.sin(xx) + np.cos(yy)
ax = plt.figure(figsize=(12,8)).add_subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap ='viridis')
plt.colorbar(p) 
```

# Contour plot
- 3D graph represented in 2D

```python
ax = plt.figure(figsize=(10,6)).add_subplot()
z = xx**2 + yy**2
p=ax.contour(xx,yy,z,cmap ='viridis')
plt.colorbar(p) 
```

```python
ax = plt.figure(figsize=(10,6)).add_subplot()
z = xx**2 + yy**2
p=ax.contourf(xx,yy,z,cmap ='viridis')
plt.colorbar(p) 
```

```python
ax = plt.figure(figsize=(10,6)).add_subplot()
z = np.sin(xx) + np.cos(yy)
p=ax.contourf(xx,yy,z,cmap ='viridis')
plt.colorbar(p) 
```
