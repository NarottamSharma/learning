# Pie Charts

```python
# simple data
data = [23,34,56,32,56]
subject = ['eng','sci','math','hin','social']
plt.pie(data , labels= subject)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
# percentage and colours
plt.pie(df['batsman_runs'], labels = df['batsman'] , autopct ='%0.1f%%')
plt.xticks(rotation = 20)
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'] , autopct ='%0.1f%%', colors =['blue','green','pink','yellow','red','orange'])
plt.xticks(rotation = 20)
plt.show()
```

```python
# Explode Shadow
plt.pie(df['batsman_runs'], labels = df['batsman'] , autopct ='%0.1f%%', explode =[0.3,0,0,0.5,0,0] , shadow = True )
plt.xticks(rotation = 20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame for illustration
data = {'batsman_runs': [12, 23, 34, 45, 56, 67, 78, 89, 90, 101, 112, 123]}
df = pd.DataFrame(data)

# Function to plot histogram with different styles
def plot_with_style(style):
    plt.style.use(style)
    plt.hist(df['batsman_runs'], bins=10)
    plt.xlabel('Runs')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Batsman Runs - {style}')
    plt.show()

# List of styles to compare
styles = ['ggplot', 'seaborn-v0_8-whitegrid', 'dark_background', 'fivethirtyeight']

for style in styles:
    plot_with_style(style)
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'] , autopct ='%0.1f%%', explode =[0.3,0,0,0.5,0,0] , shadow = True )
plt.show()
```
