```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
plt.style.use('ggplot')
```

# Scatter plot

```python
iris = pd.read_csv('iris.csv')
batters = pd.read_csv('batter.csv')
iris.sample(5)
```

```python
iris['Species'] =iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(4)
```

```python
# cmap , alpha , colorbar
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap ='viridis' ,alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# Plot size

```python
plt.figure(figsize = (15,7))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap ='viridis' ,alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# Annotation

```python
batters.head()
```

```python
sample_df = batters.head(100).sample(25 , random_state=5)
sample_df
```

```python
plt.figure(figsize = (18,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'], s=sample_df['runs'] , alpha =0.7 )

plt.axhline(130 , color = 'violet')
plt.axvline(30 , color = 'red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i] , sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'point 1')
plt.text(2,6,'point 2')
plt.text(3,7,'point 3')
plt.text(4,8,'point 4', fontdict = {'size':12,'color':'blue'})
```

# SubPlots
A different way of plotting graph

```python
# batter dataset
batters.head()
```

```python
plt.figure(figsize = (15,6))
plt.scatter(batters['avg'],batters['strike_rate'],alpha =0.7)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
# title in this 
plt.title('Something')
```

```python
fig , ax = plt.subplots(figsize = (15,6))
ax.scatter( batters['avg'] , batters['strike_rate'] ,color='pink' , marker='+')
ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike')
```

```python
fig, ax = plt.subplots()
ax.plot(batters['avg'] , batters['strike_rate'])
ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike')
```

```python
# side by side plot
plt.figure(figsize = (15,20))
plt.subplots(nrows=2, ncols=2)
```

```python
fig, ax= plt.subplots(nrows=2, ncols=1 ,sharex=True, figsize= (10,6)) # sharex is when same on x axis it will show value in 
ax[0].scatter(batters['avg'],batters['strike_rate']) # below plot because they are shared
ax[1].scatter(batters['avg'],batters['runs'])
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strike rate')
ax[1].set_title('Average vs runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Run')
plt.plot()
```
