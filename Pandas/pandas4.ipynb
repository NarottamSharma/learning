### apply

```python
# find the no of movies starting with A for each grp
def startwitha(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(startwitha)
```

```python
# find the ranking of each movie in the grp according to IMDB score
def rank_movie(group):
    group['Genre_Rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_movie).iloc[0::]
```

```python
# find normalized rating group wise
def normal(group):
    group['Normal'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/ (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal)
```

```python
# find batsman with most sixes
six = ipl[ipl['batsman_runs'] == 6]
ipl.columns
```

```python
# find batsman with most no of 4s and 5s in last 4 over
temp_df = ipl[ipl['over']>16]
```

```python
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
```

```python
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending = False).head(1).index[0]
```

```python
# find v kohli record against all team
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df
```

```python
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# create a function that can return the highest score of any batsman 
def highest(batsman):
    temp_df = ipl[ipl['batsman']==batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0]
```
